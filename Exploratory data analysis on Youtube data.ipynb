#   <font size=3 color='brown' >  11. Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences..</font>

```python
data.head(1)
```

```python
max_likes=data.pivot_table(index="title",values="likes",aggfunc="max").sort_values(by="likes",ascending=False)
print("The title of the video having max number of likes is:",max_likes.index[0],"\nThe no of likes of the video is",int(max_likes.values[0]))
max_likes.head(1)
```

```python
min_likes=data.pivot_table(index="title",values="likes",aggfunc="min").sort_values(by="likes",ascending=True)
print("The title of the video having min number of likes is:",min_likes.index[0],"\nThe no of likes of the video is",int(min_likes.values[0]))
min_likes.head(1)
```


<font size=3 color='red' > **Infernces:**</font>
<p><font size=2.5 color='black' > It can be seen that range for the likes is very large with max value as high as 31837675 and the lowest value being a low as 22 <br>it can assumed that the most popular video is "BTS () 'Dynamite' Official MV"<br>The video with min likes is "Leading the Charge | Circle K" this may also be due to lower view_counts.</font>

#   <font size=3 color='brown' >  13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.</font>

```python
data.head(1)
```

```python
data.corr()
```

```python
plt.figure(figsize=(16, 5))
ax = plt.gca()


ax.set_facecolor('Black')
plt.scatter(data["view_count"],data["dislikes"],linewidths=2,c="white",alpha=0.5)
plt.title("Dislike V/s Views Scatter Plot",fontsize=16)
plt.xlabel("Viewcount",fontsize=12)
plt.ylabel("Dislikes",fontsize=12)
plt.show()
```

<font size=3 color='red' > **Infernces:**</font>
<p><font size=2.5 color='black' >It can be seen from the scatter plot that there is a increase in the no of Dislikes with increase of view count there is a slight increase in the dislikes. <br> Also it can be infered from the correlation matrice that there is positive correlation of 0.68 between dislikes and view count so that 68% change in dislikes can be explained with change in viewcount </font>
